```python
APP_ID = dbutils.secrets.get(scope='adlgen2',key='app-id')
print(APP_ID)
SECRET_ID = dbutils.secrets.get(scope='adlgen2',key='secret-id')
print(SECRET_ID)
TENANT_ID = dbutils.secrets.get(scope='adlgen2',key='tenant-id')
print(TENANT_ID)
```

```python
spark.conf.set("fs.azure.account.auth.type.bhimstorage1.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.bhimstorage1.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.bhimstorage1.dfs.core.windows.net", APP_ID)
spark.conf.set("fs.azure.account.oauth2.client.secret.bhimstorage1.dfs.core.windows.net", SECRET_ID)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.bhimstorage1.dfs.core.windows.net", "https://login.microsoftonline.com/"+TENANT_ID+"/oauth2/token")
```
